```python
##Minimax without pruning (Takes longer.)
#Recursive function that returns the minimax score of a node
def minimax(board, player):
    global current
    current_score = check_game_over(board, player)
    
    if current_score is not False:
        return current_score
    
    scores = []
    moves = []
    p1_win = False
    p2_win = False

    #check all possible moves.
    for i in range(9):  
      if board[i] == i:
        moves.append(i)
        
    # Find the move with the highest score.   Add that move to current and return that score.            
    if player == "P1":
      best_score = -float("inf")
      best_move = moves[0]
      best_letter = ""
    
      for i in moves:
        #Choose (try "x" first)
        board[i] = "x"
        #Explore and Update Values
        current_score = minimax(board, "P2")
        if (current_score > best_score):
          best_move = i
          best_score = current_score
          best_letter ="x"
        #Unchoose
        board[i] = i

        #Choose (try "o" next)
        board[i] = "o"
        #Explore and Update Values
        current_score = minimax(board, "P2")
        if current_score > best_score:
          best_move = i
          best_score = current_score
          best_letter ="o"
        board[i] = i

      board_picked = []
      for i in range(9):
        board_picked.append(board[i])     
      board_picked[best_move] = best_letter
      current.append(board_picked)         
      return best_score
    
    
    elif player == "P2":
    # Find the move with the lowest score.   Add that move to current and return that score.    
      best_score = float("inf")
      best_move = moves[0]
      best_letter = ""
      
      #Explore and Update Values           
      for i in moves:
        #Choose (try "o" first)
        board[i] = "o"
        #Explore and Update Values
        current_score = minimax(board, "P1")
        if (current_score < best_score):
          best_move = i
          best_score = current_score
          best_letter ="o"
        #Unchoose
        board[i] = i

        #Choose (try "x" next)
        board[i] = "x"
        #Explore and Update Values
        current_score = minimax(board, "P1")
        if (current_score < best_score):
          best_move = i
          best_score = current_score
          best_letter ="x"
        board[i] = i

      board_picked = []
      for i in range(9):
        board_picked.append(board[i])     
      board_picked[best_move] = best_letter
      current.append(board_picked)         
      #board[best_move] = best_move  #Return board back to normal to continue exploring
      return best_score
```

```python
#Minimax with Alpha/Beta Pruning
def minimax(board, player, alpha, beta):
    global current
    current_score = check_game_over(board, player)

    if current_score is not False:
        return current_score
    
    moves = []
    p1_win = False
    p2_win = False

    #check all possible moves.
    for i in range(9):  
        if board[i] == i:
            moves.append(i)
        
    # Find the move with the highest/lowest score. Add that move to current and return that score.
    if player == "P1":
        best_score = -float("inf")
        best_move = moves[0]
        best_letter = ""

        for i in moves:
            # Choose (try "x" first)
            board[i] = "x"
            # Explore and Update Values
            current_score = minimax(board, "P2", alpha, beta)
            if current_score > best_score:
                best_move = i
                best_score = current_score
                best_letter = "x"
            # Unchoose
            board[i] = i

            alpha = max(alpha, best_score)
            if beta <= alpha:
                break

            # Choose (try "o" next)
            board[i] = "o"
            # Explore and Update Values
            current_score = minimax(board, "P2", alpha, beta)
            if current_score > best_score:
                best_move = i
                best_score = current_score
                best_letter = "o"
            board[i] = i

            alpha = max(alpha, best_score)
            if beta <= alpha:
                break

        board_picked = []
        for i in range(9):
            board_picked.append(board[i])
        board_picked[best_move] = best_letter
        current.append(board_picked)
        return best_score


    elif player == "P2":
        # Find the move with the lowest score. Add that move to current and return that score.
        best_score = float("inf")
        best_move = moves[0]
        best_letter = ""

        for i in moves:
            # Choose (try "o" first)
            board[i] = "o"
            # Explore and Update Values
            current_score = minimax(board, "P1", alpha, beta)
            if current_score < best_score:
                best_move = i
                best_score = current_score
                best_letter = "o"
            # Unchoose
            board[i] = i

            beta = min(beta, best_score)
            if beta <= alpha:
                break

            # Choose (try "x" next)
            board[i] = "x"
            # Explore and Update Values
            current_score = minimax(board, "P1", alpha, beta)
            if current_score < best_score:
                best_move = i
                best_score = current_score
                best_letter = "x"
            board[i] = i

            beta = min(beta, best_score)
            if beta <= alpha:
                break

        board_picked = []
        for i in range(9):
            board_picked.append(board[i])
        board_picked[best_move] = best_letter
        current.append(board_picked)
        return best_score
```
